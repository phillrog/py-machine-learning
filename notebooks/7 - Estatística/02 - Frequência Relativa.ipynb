```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def calcula_ticks(lista,barras):
    lista.sort()
    menor_valor = lista[0]
    maior_valor = lista[-1]
    intervalo = (maior_valor - menor_valor)/barras
    ticks = [menor_valor]
    ultimo = menor_valor+intervalo
    ticks.append(ultimo)
    for x in range(barras-1):
        ultimo += intervalo
        ticks.append(ultimo)
    return ticks
```

```python
km = pd.Series([4,6,6,7,11,13,18,19,21,24,26,27,35,36,36,42,43,45,49])
```

```python
lista = km.values
barras = 5
km.plot.hist(bins=barras, rwidth=0.95)
ticks = calcula_ticks(lista, barras)
plt.xticks(ticks)
plt.title('Distância de Casa para Trabalho')
plt.xlabel('km')
plt.ylabel('Frequência')
plt.grid(axis='y')
```

```python
def frequencia_relativa(total, ticks):
    frequencia_relati = []
    for item in ticks:
        x = item / total
        frequencia_relati.append("{0:.2f}%".format(x*100))
    return frequencia_relati
```

```python
lista = km.values
barras = 5
ax = km.plot.hist(bins=barras, rwidth=0.95)
yticks = ax.get_yticks()
total = len(lista)
freq = frequencia_relativa(total, yticks)
ax.set_yticklabels(freq)
ticks = calcula_ticks(lista, barras)
plt.xticks(ticks)
plt.title('Distância de Casa para Trabalho')
plt.xlabel('km')
plt.ylabel('Frequência')
plt.grid(axis='y')
```
