```python
def calcula_ticks(lista, barras):
    maior_valor = lista[0]
    menor_valor = lista[-1]
    intervalo = (maior_valor-menor_valor)/barras
    ticks = [menor_valor]
    ultimo = menor_valor+intervalo
    ticks.append(ultimo)
    
    for x in range(barras-1):
        ultimo += intervalo
        ticks.append(ultimo)
    return ticks
    
```

```python
lista = km.values
barras=10
km.plot.hist(bins=barras, rwidth=0.95)
calcula_ticks(lista, barras)
plt.xticks(ticks)
plt.title('Distância de Casa para Trabalho')
plt.xlabel('Km')
plt.ylabel('Frequência')
plt.grid(axis='y')
```
